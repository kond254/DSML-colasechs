```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
import seaborn as sns
from datetime import date, time, datetime, timedelta 
```

```python
Bosten_set=pd.read_csv("boston_2015.csv")
```

```python
Bosten_set.sort_values(["start_time","bike_id"], inplace=True)
Bosten_set.head(10)
```

```python
Bosten_set["start_time"] = pd.to_datetime(Bosten_set["start_time"])
Bosten_set["end_time"] = pd.to_datetime(Bosten_set["end_time"])
#Bosten_set["end_station_id"].astype("int")
#Bosten_set["end_station_id"]=pd.DataFrame(Bosten_set["end_station_id"],dtype=np.int)
```

```python
Bosten_set.info()
```

```python
Bosten_set["Duration"]=Bosten_set.end_time-Bosten_set.start_time
```

```python
Bosten_set.head()
```

```python
def get_date (ts):
    return ts.date()

def get_weekday (ts):
    return ts.weekday()

def get_hour (ts):
    return ts.hour
```

```python
Bosten_set.reset_index(inplace=True)
Bosten_set["Date"] = Bosten_set["start_time"].apply(lambda ts: get_date (ts))
Bosten_set["Weekday"]=Bosten_set["start_time"].apply(lambda x: get_weekday (x))
Bosten_set["Hour"]=Bosten_set["start_time"].apply(lambda x: get_hour (x))
```

```python
Bosten_set.head()
```

```python
Bosten_set.describe()
```

```python
Bosten_set_cl=Bosten_set[Bosten_set['Duration'] >= pd.Timedelta(0)]
```

```python
#Bosten_set_clean = Bosten_set_clean[Bosten_set_clean['Duration'].dt.days <=0]
```

```python
Bosten_set_clean=Bosten_set_cl[Bosten_set_cl['Duration'] <= pd.Timedelta("1 day")]
```

```python
Bosten_set_clean.describe()
```

```python
len(Bosten_set_clean)-len(Bosten_set) # delete 17 colomn where duration <0 and duration >1day
```

```python
Bosten_set_clean_avail = Bosten_set_clean.groupby(["Date","Weekday"])["bike_id"].nunique()
Bosten_set_clean_avail = pd.DataFrame(Bosten_set_clean_avail)
```

```python
fig,ax = plt.subplots(figsize=(10,4))  #(0 = Monday, 6 = Sunday)

sns.boxplot(x=Bosten_set_clean_avail.index.get_level_values(1), y=Bosten_set_clean_avail["bike_id"],ax=ax)
plt.show()
```

```python
Bosten_set_clean_avail = Bosten_set_clean.groupby(["Date","Hour"])["bike_id"].nunique()
Bosten_set_clean_avail = pd.DataFrame(Bosten_set_clean_avail)
```

```python
fig,ax = plt.subplots(figsize=(10,4))  

sns.boxplot(x=Bosten_set_clean_avail.index.get_level_values(1), y=Bosten_set_clean_avail["bike_id"],ax=ax)
plt.show()
```
