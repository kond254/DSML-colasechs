```python
import numpy as np
import pandas as pd
from datetime import date, time, datetime, timedelta 
from dateutil import parser

import warnings
warnings.filterwarnings("ignore")
```

# Creation of our featureset

```python
mydateparser = lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S")
df_boston=pd.read_csv("Boston/boston_2015.csv", parse_dates=['start_time', "end_time"], date_parser=mydateparser)
```

```python
df_boston = df_boston.drop(df_boston.index[21566])
```

```python
df_weather=pd.read_csv("Boston/weather_hourly_boston.csv")
```

```python
df_weather.info()
```

```python
df_weather = df_weather.dropna()
df_weather.reset_index(drop=True, inplace=True)
df_weather.info()
```

```python
df_weather["date_time"]=pd.to_datetime(df_weather["date_time"])
```

```python
df_weather
```

```python
df_weather.sort_values("date_time",ignore_index=True, inplace=True,ascending=True)
df_weather
```

```python
only_2015=datetime.strptime("2016-01-01", "%Y-%m-%d")
only_2015
```

```python
df_weather=df_weather[df_weather["date_time"]<only_2015]
```

```python
df_boston=df_boston.drop(columns=["start_station_id", "end_station_id","start_station_name","end_station_name","bike_id","user_type","end_time"])
```

```python
df_boston
```

```python
df_weather["mean_temp"]=(df_weather["max_temp"]+df_weather["min_temp"])/2
```

```python
df_boston=df_boston.dropna()
df_boston.reset_index(drop=True, inplace=True)
```

```python
df_boston.info()
```

```python
new_df=pd.DataFrame(columns=["start_time","count","mean_temp","precip"])
new_df
```

```python
#creation of df in which only days and hours are displayed
p_time=datetime.strptime("2015-01-01 00:00:00", "%Y-%m-%d %H:%M:%S")
for i in range(24*365):
    df_akt=df_boston[(df_boston["start_time"]>=p_time) & (df_boston["start_time"]<p_time+timedelta(hours=1))]
    df_akt=df_akt.reset_index()
    count=mean=precip=None
    if len(df_akt) is not 0:
        #mean=df_akt["mean_temp"][0]
        count=len(df_akt)
        #precip=df_akt["precip"][0]
    #akt_dict={"start_time":p_time,"count":count,"mean_temp":mean,"precip":precip}
    akt_dict={"start_time":p_time,"count":count}
    new_df=new_df.append(akt_dict,ignore_index=True)
    p_time=p_time+timedelta(hours=1)
    count=mean=precip=None
```

```python
new_df
```

```python
p=0
for i in range(len(new_df)):
    time=new_df["start_time"][i]
    if len(df_weather[df_weather["date_time"]==time]) is not 0:
        new_df["mean_temp"][i]=df_weather["mean_temp"][p]
        new_df["precip"][i]=df_weather["precip"][p]
        p=p+1
```

```python
new_df.dropna(subset = ["mean_temp"], inplace=True)
```

```python
#filling values where there is weather data but no trips have been recorded
new_df["count"]=new_df["count"].fillna(0)
```

```python
new_df[0:15]
```

```python
new_df["spring"]=new_df["start_time"].apply(lambda x: 1 if x.month==3 or x.month==4 or x.month==5 else 0)
new_df["summer"]=new_df["start_time"].apply(lambda x: 1 if x.month==6 or x.month==7 or x.month==8 else 0)
new_df["autumn"]=new_df["start_time"].apply(lambda x: 1 if x.month==9 or x.month==10 or x.month==11 else 0)
new_df["winter"]=new_df["start_time"].apply(lambda x: 1 if x.month==12 or x.month==1 or x.month==2 else 0)
```

```python
new_df
```

```python
new_df['rushhour'] = new_df['start_time'].apply(lambda x: 1 if (7<=x.hour<=9) or (16<=x.hour<=18)  else 0)
```

```python
new_df
```

```python
new_df["precip"]=new_df["precip"].astype(int)
new_df.info()
```

```python
def get_weekday (ts):
    return ts.weekday()
#0=Monday 6=Sunday
new_df["weekday"]=new_df["start_time"].apply(lambda x:1 if 0<=get_weekday(x)<5 else 0)
```

```python
new_df["season"]=new_df["start_time"].apply(lambda x: 4 if x.month==12 or x.month==1 or x.month==2 else 
                                                      1 if x.month==3 or x.month==4 or x.month==5 else 
                                                      2 if x.month==6 or x.month==7 or x.month==8 else 3)
```

```python
new_df["hour"]=new_df["start_time"].apply(lambda x: x.hour)
```

```python
new_df.info()
```

```python
df_boston=new_df.copy()
```

```python
df_boston["mean_temp"]=df_boston["mean_temp"].astype(float)
```
